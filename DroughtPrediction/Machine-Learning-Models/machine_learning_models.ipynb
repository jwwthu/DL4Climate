```python
import numpy as np
import pandas as pd
from pycaret.classification import *
```

```python
input_fn = './data_ndarray.npz'
```

```python
print(input_fn)
```

```python
data = np.load(open(input_fn, 'rb'))
```

```python
X_train = data['X_train']
```

```python
print(X_train.shape)
```

```python
X_valid = data['X_valid']
```

```python
print(X_valid.shape)
```

```python
X_test = data['X_test']
```

```python
print(X_test.shape)
```

```python
y_train = data['y_train']
```

```python
y_train = np.array([int(i) for i in y_train])
```

```python
print(y_train.shape)
```

```python
y_valid = data['y_valid']
```

```python
y_valid = np.array([int(i) for i in y_valid])
```

```python
print(y_valid.shape)
```

```python
y_test = data['y_test']
```

```python
y_test = np.array([int(i) for i in y_test])
```

```python
print(y_test.shape)
```

```python
X_train = X_train.reshape(X_train.shape[0], -1)
```

```python
print(X_train.shape)
```

```python
X_valid = X_valid.reshape(X_valid.shape[0], -1)
```

```python
print(X_valid.shape)
```

```python
X_test = X_test.reshape(X_test.shape[0], -1)
```

```python
print(X_test.shape)
```

Concat the data:

```python
X = np.concatenate([X_train, X_valid], axis=0)
```

```python
X.shape
```

```python
y = np.concatenate([y_train, y_valid], axis=0)
```

```python
y.shape
```

Build the data into the DataFrame format as required by pycaret:

```python
numeric_features = ['f' + str(i) for i in range(X_train.shape[1])]
target_col = 'label'
df = pd.DataFrame(X, columns=numeric_features)
df[target_col] = y
df.head()
```

```python
numeric_features = ['f' + str(i) for i in range(X_train.shape[1])]
target_col = 'label'
df_test = pd.DataFrame(X_test, columns=numeric_features)
df_test[target_col] = y_test
df_test.head()
```

Set the random seed:

```python
random_seed = 42
```

```python
exp_clf = setup(data=df,
                session_id=random_seed,
                numeric_features=numeric_features,
                target=target_col,
                normalize = False,  # the data is already z-score normalized
                test_data = df_test,  # test data
                silent=True)
```

```python
import time
```

# 1. Logistic regression

```python
start_time = time.time()
```

```python
model = create_model('lr')
```

```python
predict_model(model);
```

```python
tuned_model= tune_model(model, fold=5, n_iter=10)
```

```python
end_time = time.time()
print(end_time - start_time)
```

```python
print(tuned_model)
```

```python
results = predict_model(tuned_model);
```

Save the true and predicted labels:

```python
np.save('y_true_lr', results['label'].values)
np.save('y_pred_lr', results['Label'].values)
```
